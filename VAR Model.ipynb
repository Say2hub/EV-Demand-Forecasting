##STATIONARITY TESTS

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title=''):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] < 0.05:
        print(f'{title} is stationary')
    else:
        print(f'{title} is not stationary')


for col in df.columns[1:]:
  # Test for stationarity of EV sales
  adf_test(df[col], title=col)
```

```python
# First difference for most variables
df_diff1 = df[['DIESEL', 'petrol', 'hybrid', 'total']].diff(1).dropna()

# Second difference for EV and CNG
df_diff2_ev_cng = df[['ev', 'cng']].diff().diff().dropna()

# Combine the differenced data into a new DataFrame
df_stationary = pd.concat([df['Date'], df_diff1, df_diff2_ev_cng], axis=1).dropna()
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title=''):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] < 0.05:
        print(f'{title} is stationary')
    else:
        print(f'{title} is not stationary')


for col in df_stationary.columns:
  # Test for stationarity of EV sales
  adf_test(df_stationary[col], title=col)
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
cols = ['DIESEL', 'petrol', 'cng', 'hybrid']
for col in cols:
  print(col)
  grangercausalitytests(df_stationary[['ev', col]], maxlag=4)

```

1. DIESEL
There is no Granger causality detected between DIESEL and ev up to 4 lags.
The p-values for all the lags are well above the significance threshold (typically 0.05), indicating that DIESEL does not Granger-cause ev.
2. Petrol
Granger causality is detected at lag 3.
At lag 3, the p-value for the F-test is 0.0176, which is below 0.05, indicating significant Granger causality.
The other test statistics (ssr-based chi2 and likelihood ratio test) also confirm this with low p-values.
At lag 4, the p-values approach significance, but they are slightly above 0.05 (e.g., 0.0738 for the F-test, 0.0502 for chi2 test), so we cannot confirm strong causality at lag 4.
3. CNG
There is strong Granger causality detected between CNG and ev across all lags.
At lag 1, the p-value for the F-test is 0.0052, which is significant.
At lag 2, the p-value drops to 0.0000, which indicates extremely strong causality.
The trend continues at lags 3 and 4 with p-values remaining at 0.0000, indicating that CNG is a significant Granger cause of ev.
4. Hybrid
Granger causality is detected at lag 3.
The p-value for the F-test is 0.0000, which is highly significant.
Granger causality is also detected at lag 4, with p-values for all tests being 0.0000, indicating strong causality at higher lags as well.
#Summary of Inferences:
###DIESEL does not appear to Granger-cause ev.
###Petrol Granger-causes ev at lag 3.
###CNG has strong causality for ev across all lags (1 to 4), with lag 2 showing the most robust causality.
###Hybrid vehicles Granger-cause ev starting at lag 3 and continuing strongly at lag 4.
*This suggests that CNG and Hybrid variables have a much stronger temporal influence on ev, while Petrol shows a moderate effect at lag 3. Diesel shows no significant influence.*

###VAR MODEL

```python
df_stationary.drop(columns = ['total'], inplace = True)
```

```python
# Convert the 'date' column to datetime format
df_stationary['Date'] = pd.to_datetime(df_stationary['Date'])

# Set the 'date' column as the index
df_stationary.set_index('Date', inplace=True)

# Verify the conversion
print(df_stationary.head())

df_stationary = df_stationary.asfreq('MS')
print(df_stationary.index.freq)
```

```python
df_stationary.drop(columns = ['DIESEL'], inplace = True)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=min(df_stationary.shape[1], 4))
data_pca = pd.DataFrame(pca.fit_transform(df_stationary), index=df_stationary.index)
```

```python
from statsmodels.tsa.api import VAR

# Fit the VAR model on differenced data
model = VAR(data_pca)

# Choose the optimal lag length using Akaike Information Criterion (AIC)
lag_order = model.select_order(maxlags=3)
print(lag_order.summary())

# Fit the model using the chosen lag length
var_model = model.fit(lag_order.aic)
print(var_model.summary())
```

```python
from statsmodels.stats.stattools import durbin_watson

# Compute Durbin-Watson statistics
dw_stats = durbin_watson(var_model.resid)

for col, stat in zip(df_stationary.columns, dw_stats):
    print(f'{col}: {stat}')
```

##2. VAR Model Selection:
We used Vector AutoRegression (VAR) to model the interactions between these variables.
We ran model selection criteria (AIC, BIC, HQIC, FPE) and determined that the best model had 3 lags based on the minimum values for each criterion.
##3. VAR Model Estimation:
The VAR model was estimated with 3 lags.
We found significant relationships in some variables, such as Petrol and Hybrid, which had notable coefficients in their relationship to EV.

##CNG and Hybrid: Both seem to have a strong influence on the EV market, suggesting competitive or complementary dynamics.
##Petrol: Influences EV adoption but with a delayed impact (3 periods lag).
##Diesel: No significant effect on the EV market.

```python
import pandas as pd
import matplotlib.pyplot as plt

def forecast_var(var_results, steps=12):
    try:
        # Get the original data
        original_data = var_results.model.data.orig_endog

        # Ensure the index is a DatetimeIndex
        if not isinstance(original_data.index, pd.DatetimeIndex):
            raise ValueError("The index must be a DatetimeIndex.")

        # Generate forecast
        forecast_input = original_data.values[-var_results.k_ar:]
        forecast = var_results.forecast(forecast_input, steps=steps)

        # Create a DataFrame with the forecasted values
        last_date = original_data.index[-1]
        forecast_index = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=steps, freq='MS')
        forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=original_data.columns)

        # Plot the forecast
        fig, axes = plt.subplots(nrows=len(original_data.columns), ncols=1, figsize=(12, 4*len(original_data.columns)))

        if len(original_data.columns) == 1:
            axes = [axes]

        for i, col in enumerate(original_data.columns):
            axes[i].plot(original_data.index, original_data[col], label='Observed')
            axes[i].plot(forecast_df.index, forecast_df[col], label='Forecast')
            axes[i].set_title(f'Forecast for {col}')
            axes[i].legend()

        plt.tight_layout()
        plt.show()

        return forecast_df

    except Exception as e:
        print(f"An error occurred: {e}")

# Usage:
forecast_results = forecast_var(var_model)
print(forecast_results)
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose

def reconstruct_var_forecast(forecast_df, pca, original_df, diff_order):
    # Step 1: Inverse PCA transformation
    inverse_pca = pca.inverse_transform(forecast_df)
    inverse_pca_df = pd.DataFrame(inverse_pca, index=forecast_df.index, columns=original_df.columns)

    # Step 2: Undo differencing
    last_original_values = original_df.iloc[-diff_order:]
    reconstructed_df = pd.DataFrame(index=forecast_df.index, columns=original_df.columns)

    for col in inverse_pca_df.columns:
        print(f"Processing column: {col}")

        # Check for NaN values
        if original_df[col].isnull().any():
            print(f"Warning: NaN values found in {col}. Interpolating...")
            original_df[col] = original_df[col].interpolate()

        # Decompose the series
        decomposition = seasonal_decompose(original_df[col], model='additive', period=12)
        trend = decomposition.trend
        seasonal = decomposition.seasonal

        # Forecast trend using simple exponential smoothing
        trend_model = ExponentialSmoothing(trend.dropna(), trend='add', seasonal=None)
        trend_fit = trend_model.fit()
        trend_forecast = trend_fit.forecast(steps=len(forecast_df))

        # Undo differencing and add back trend and seasonality
        if diff_order == 1:
            values = inverse_pca_df[col].cumsum() + last_original_values[col].iloc[-1]
        elif diff_order == 2:
            temp = inverse_pca_df[col].cumsum() + last_original_values[col].diff().iloc[-1]
            values = temp.cumsum() + last_original_values[col].iloc[-1]

        # Add seasonality
        seasonal_pattern = np.tile(seasonal[-12:], len(forecast_df)//12 + 1)[:len(forecast_df)]

        reconstructed_df[col] = values + trend_forecast + seasonal_pattern

    return reconstructed_df

# Usage
try:
    reconstructed_forecast = reconstruct_var_forecast(forecast_df, pca, df, diff_order=2)

    # Plotting for EV sales
    plt.figure(figsize=(12, 6))
    plt.plot(df.index[-100:], df['ev'][-100:], label='Historical EV Sales', color='blue')
    plt.plot(reconstructed_forecast.index, reconstructed_forecast['ev'], label='Forecasted EV Sales', color='orange', marker='*')
    plt.title('Historical and Predicted EV Sales')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.legend()
    plt.show()

    print(reconstructed_forecast)
except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("Please check your data for any inconsistencies.")
```
