##STATIONARITY TESTS

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title=''):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] < 0.05:
        print(f'{title} is stationary')
    else:
        print(f'{title} is not stationary')


for col in df.columns[1:]:
  # Test for stationarity of EV sales
  adf_test(df[col], title=col)
```

```python
# First difference for most variables
df_diff1 = df[['DIESEL', 'petrol', 'hybrid', 'total']].diff(1).dropna()

# Second difference for EV and CNG
df_diff2_ev_cng = df[['ev', 'cng']].diff().diff().dropna()

# Combine the differenced data into a new DataFrame
df_stationary = pd.concat([df['Date'], df_diff1, df_diff2_ev_cng], axis=1).dropna()
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title=''):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] < 0.05:
        print(f'{title} is stationary')
    else:
        print(f'{title} is not stationary')


for col in df_stationary.columns:
  # Test for stationarity of EV sales
  adf_test(df_stationary[col], title=col)
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
cols = ['DIESEL', 'petrol', 'cng', 'hybrid']
for col in cols:
  print(col)
  grangercausalitytests(df_stationary[['ev', col]], maxlag=4)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
diesel_series = df_stationary['DIESEL']
plt.figure(figsize=(12, 6))

# ACF plot-q
plt.subplot(121)
plot_acf(diesel_series, lags=40, ax=plt.gca())
plt.title('ACF Plot')

# PACF plot-p
plt.subplot(122)
plot_pacf(diesel_series, lags=40, ax=plt.gca())
plt.title('PACF Plot')

plt.tight_layout()
plt.show()
```

```python
if 'Date' not in df.columns:
    df = df.reset_index()
    if df.columns[0] != 'Date':
        df = df.rename(columns={df.columns[0]: 'Date'})

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.asfreq('MS')
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(df_stationary['DIESEL'],
              exog=df_stationary[['ev', 'hybrid', 'cng', 'petrol']],
              order=(1, 0, 1))  # Setting d=0 for ARMA
model_fit = model.fit(method_kwargs={"maxiter": 500})

# Display the model summary
print(model_fit.summary())
```

```python
forecast_periods = 12
exog_forecast = df_stationary[['ev', 'hybrid', 'cng', 'petrol']].iloc[-forecast_periods:]
forecast = model_fit.forecast(steps=forecast_periods, exog=exog_forecast)
```

```python
train_size = int(len(df_stationary) * 0.8)
train = df_stationary[:train_size]
test = df_stationary[train_size:]
```

```python
forecast_index = range(len(train) + len(test), len(train) + len(test) + forecast_periods)
plt.plot(train.index, train['DIESEL'], label='Train Data')
plt.plot(test.index, test['DIESEL'], label='Test Data')
plt.plot(forecast_index, forecast, label='Forecast', color='red')
plt.legend()
plt.title('ARMA Forecast')
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

residuals = model_fit.resid

# Histogram with density curve
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, density=True, alpha=0.6, color='b')
mean, std = residuals.mean(), residuals.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Histogram of Residuals')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
forecast_values = forecast.loc[forecast.index.isin(test['Date'])]
target_column = 'DIESEL'
mse = mean_squared_error(test[target_column], forecast_values)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((test[target_column] - forecast_values) / test[target_column])) * 100
r2 = r2_score(test[target_column], forecast_values)

print(f'MSE: {mse}, RMSE: {rmse}, MAPE: {mape}, R²: {r2}')
```

```python
exog_vars = ['ev', 'hybrid', 'cng', 'petrol']

forecast = model_fit.forecast(steps=len(test), exog=test[exog_vars])

plt.figure(figsize=(10, 6))
plt.plot(train.index, train['ev'], label='Train', color='blue')
plt.plot(test.index, test['ev'], label='Test', color='green')
plt.plot(test.index, forecast, label='Forecast', color='red')

plt.title('Forecast for EV with 95% Confidence Interval')
plt.legend()
plt.show()
```
