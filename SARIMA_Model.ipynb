##STATIONARITY TESTS

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series, title=''):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4f}')
    if result[1] < 0.05:
        print(f'{title} is stationary\n')
    else:
        print(f'{title} is not stationary\n')


for col in df.columns[1:]:
  # Test for stationarity of EV sales
  adf_test(df[col], title=col)
```

```python
if 'Date' not in df.columns:
    df = df.reset_index()
    if df.columns[0] != 'Date':
        df = df.rename(columns={df.columns[0]: 'Date'})

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.asfreq('MS')
```

```python
train_size = int(len(df) * 0.8)
train = df[:train_size]
test = df[train_size:]
```

```python
from sklearn.preprocessing import StandardScaler
target = 'ev'
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train[target].values.reshape(-1, 1))
```

```python
test_scaled = scaler.transform(test[target].values.reshape(-1, 1))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
exog_vars = ['DIESEL', 'petrol', 'hybrid', 'total']
try:
    model = SARIMAX(train_scaled,
                    order=(0, 1, 1),
                    seasonal_order=(0, 0, 1, 12),
                    exog=train[exog_vars],
                    enforce_stationarity=False,
                    enforce_invertibility=False)

    model_fit = model.fit(maxiter=500, disp=True)
    print(model_fit.summary())
except Exception as e:
    print(f"Error fitting the model: {e}")
```

```python
forecast_scaled = model_fit.predict(start=len(train_scaled),
                                    end=len(train_scaled) + len(test_scaled) - 1,
                                    exog=test[exog_vars])
```

```python
forecast = scaler.inverse_transform(forecast_scaled.to_numpy().reshape(-1, 1))
test_actual = scaler.inverse_transform(test_scaled.reshape(-1, 1))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
plt.plot(test.index, test_actual, label='Actual Data', color='blue')
plt.plot(test.index, forecast, label='Forecast', color='red')
plt.title('EV Sales Forecast')
plt.xlabel('Date')
plt.ylabel('EV Sales')
plt.legend()
plt.show()
forecast_std_err = np.std(model_fit.resid)
ci_lower = forecast.flatten() - 1.96 * forecast_std_err
ci_upper = forecast.flatten() + 1.96 * forecast_std_err

plt.fill_between(test.index,
                 ci_lower,
                 ci_upper,
                 color='gray', alpha=0.3, label='95% CI')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

residuals = model_fit.resid

# Histogram with density curve
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, density=True, alpha=0.6, color='b')
mean, std = residuals.mean(), residuals.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Histogram of Residuals')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error

mse = mean_squared_error(test_actual, forecast)
rmse = mse ** 0.5
mape = mean_absolute_percentage_error(test_actual, forecast) * 100
r2 = r2_score(test_actual, forecast)

print(f'MSE: {mse}, RMSE: {rmse}, MAPE: {mape}%, R²: {r2}')
```

```python
standardized_residuals = (residuals - residuals.mean()) / residuals.std()
plt.plot(standardized_residuals)
plt.title('Standardized Residuals')
plt.show()
```
